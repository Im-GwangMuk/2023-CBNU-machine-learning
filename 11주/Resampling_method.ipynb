```python
import pandas as pd
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler
from collections import Counter
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/drive/MyDrive/기계학습 실습 /11주/creditcard.csv")
```

```python
y= data.pop("Class")
X = data
```

```python
print(sorted(Counter(y).items()))
```

# Oversampling

```python
ros = RandomOverSampler()
X_resampled, y_resampled = ros.fit_resample(X, y)
```

```python
print(sorted(Counter(y_resampled).items()))
```

```python
class_label_1_over = X_resampled[y_resampled == 1]
class_label_0_over = X_resampled[y_resampled == 0]
```

```python
plt.scatter(class_label_1_over["V1"],class_label_1_over["V2"],color="red")
plt.scatter(class_label_0_over["V1"],class_label_0_over["V2"],color="blue")
plt.xlabel("V1")
plt.ylabel("V2")
```

# Undersampling

```python
rus = RandomUnderSampler()
X_resampled, y_resampled = rus.fit_resample(X, y)
```

```python
print(sorted(Counter(y_resampled).items()))
```

```python
class_label_1_under = X_resampled[y_resampled == 1]
class_label_0_under = X_resampled[y_resampled == 0]
```

```python
plt.scatter(class_label_1_under["V1"],class_label_1_under["V2"],color="red")
plt.scatter(class_label_0_under["V1"],class_label_0_under["V2"],color="blue")
plt.xlabel("V1")
plt.ylabel("V2")
```
