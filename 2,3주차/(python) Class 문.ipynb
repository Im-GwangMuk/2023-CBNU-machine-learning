```python
class JSS :
  def __init__(self) :                  # __init__  :  클래스를 선언하는 순간 실행되는 함수.
    print("JSS class 선언!")            # self : 클래스를 저장할 변수

  def show(self) : 
    print("show 실행!")
```

## **동빈나**


- **클래스 (Class)**

  반복되는 불필요한 소스코드를 최소화 하면서 현실 세계의 사물을 \
컴퓨터 프로그래밍 상에서 쉽게 표현할 수 있도록 해주는 프로그래밍 기술.

  - 클래스의 멤버 : 클래스 내부에 포함되는 변수.
  - 클래스의 함수 : 클래스 내부에 포함되는 함수.  메소드라고 부름.

\

- **인스턴스**

  클래스로 정의된 객체를 프로그램 상에서 이용할 수 있게 만든 변수


### **기초**

```python
class Car :
  # 클래스의 생성자
  def __init__(self, name, color) :
    self.name = name      # 클래스의 멤버
    self.color = color    # 클래스의 멤버


  # 클래스의 메소드
  def show_info(self) :
    print("이름 : {}  /  색상 : {}".format(self.name, self.color))


  # Seter 메소드 - 특정한 값을 변경시 사용.
  def set_name(self, name) :
    self.name = name


  # 클래스 소멸자 - 클래스 메모리를 삭제 시 사용.
  def __del__(self) :
    print("인스턴스를 소멸시킴!")
```

```python
car1 = Car("소나타", "빨간색")
car1.show_info()
```

```python
car2 = Car("아반떼", "검은색")
car2.show_info()
```

```python
print(car1.name , "을(를) 구매하였습니다!")
print(car2.color , "으로 구매하였습니다!")
```

```python
car1 = Car("소나타", "빨간색")
car1.set_name("티볼리")
car1.show_info()
```

```python
car1 = Car("소나타", "빨간색")
car1.set_name("티볼리")
car1.show_info()
del car1       # car1을 메모리 상에서 제거.

# car1.show_info()    # car1을 메모리 상에서 지웠으므로 호출 불가.
```

### **상속**

- 다른 클래스의 멤버 변수와 메소드를 받아 사용하는 기법.

- 만약 자식클래스의 메소드 이름이 부모 클래스의 메소드와 동일하다면, 자식클래스의 메소드가 우선.

```python
# 부모 클래스
class Unit :
  def __init__(self, name, power) :
    self.name = name
    self.power = power

  def attack(self) :
    print("{} 이(가) 공격을 수행하였습니다. [전투력 : {}]".format(self.name, self.power))
```

```python
unit = Unit("홍길동", 500)
unit.attack()
```

```python
class Monster(Unit) :
  def __init__(self, name, power, type) :
    self.name = name
    self.power = power
    self.type = type

  def show_info(self) :
    print("몬스터 이름 : {} / 몬스터 종류 : {}".format(self.name, self.type))
```

```python
monster = Monster("슬라임", 30, "액괴")
monster.attack()
monster.show_info()
```

## **테디노트**

- 부모 클래스 (Super class) 　:　 모태가 되는 클래스

- 자식 클래스 (Sub class, Child class) 　:　 부모로 부터 상속 받은 클래스

- 생성자 (Constructor) 　:　  클래스 인스턴스를 생성

- __ init __() 메서드 　:　 클래스 생성자 메서드

- 클래스 변수 　:　 클래스에서 공유 되는 변수

- 메서드 (Method)

- 메서드 오버라이딩 (Method Overriding) 　:　 상속 받은 메서드를 재정의.


### **기초**

```python
class Person_age :

  total_count = 0                # 클래스 변수 : 클래스에서 공유되는 변수

  def __init__(self) :           # 생성자 메서드
    self.name = "홍길동"
    self.age = 3
    Person.total_count += 1


  def introduce(self) :          # 메서드
    print("제 이름은 {}이고, 나이는 {}살 입니다.".format(self.name, self.age))
```

### **상속**

```python
class Student(Person) :

  def __init__(self) :
    super().__init__("콩쥐")      # 부모 클래스의 __init__ 함수를 호출.
```

```python
class Student(Person) :

  def __init__(self, name) :
    super().__init__(name)        # 부모 클래스의 __init__ 함수를 호출.

  def introduce(self) :           # 메서드 이름이 같은 경우 자식 클래스가 우선.
    super().introduce()           # 주석 처리 시 자식 클래스의 메서드만 출력 됨.
    print("(자식 클래스) 제 나이는 {}살 이고, 이름은 {} 입니다.".format(self.age, self.name))
```

```python
s1 = Student("콩쥐")
s1.introduce()
```

## **self 관련**

```python
class Person :

  total_count = 0                
  
  def __init__(self, age, ages) :       
    self.age = age                      # self. 를 붙이면 class 내 메소드(def)에서 파라미터 호출없이 사용 가능.
    ages = ages                         # self. 를 "안"붙이면 class 내 메소드(def)에서 파라미터 호출없이 사용 "불"가능.
    Person.total_count += 1


  def introduce(self, ages) :          # ages는 위에서 self.을 안했으므로 다시 호출해줘야 함.  즉, def introduce(self) 시 에러.
    self.age += 1
    print("나이는 {}살 이고 만나이는 {}살 입니다.".format(ages, self.age)) 
    

  def introduce2(ages) :              # def name (self)를 넣지 않은 경우 호출 시 +1 개의 인자가 자동으로 더 들어감 따라서 인자 개수가 안 맞음.
                                      # 여기선 인자가 1개만 필요한데 self가 없어서 필수 인자 1개(ages) 자동으로 1개 해서 총 2개가 들어가므로 오류가 남.
    print("나이는 {}살 이고 만나이는 {}살 입니다.".format(ages, ages))
    
```

```python
a = Person(5, 5)

print(a.introduce(5))
#print("\n==================\n")
#print(a.introduce2(5))
```
